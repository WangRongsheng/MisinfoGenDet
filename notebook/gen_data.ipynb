```python
import json
import time
from tqdm import tqdm

'''read json'''
with open('dev.json') as f:
    # read json
    data = json.load(f)

'''use json to generate fake data'''
results = []
results_fake = []
count = 0
#for i in tqdm(range(len(data))):
for i in tqdm(range(50)):
    if data[i]['label']=="supports":
        count += 1
        time.sleep(5)
        response = openai.ChatCompletion.create(
                    model="gpt-3.5-turbo",
                    messages=[{"role": "user", "content": "将以下内容改为虚假的信息，你可以在以下改写方法中任意选择一种合适的方法：1. 置换这句话中两个人物的身份；2.在原来的内容的基础上为他们随意添加一些经典的著作；3.涉及数字的内容你可以直接修改数字信息；\n\n你需要改写的句子是："+str(data[i]['claim'])}],
                        temperature=0.7,
                        max_tokens=1024
                    )

        info_fake = {
            "context": str(response.choices[0].message.content)
        }
        results_fake.append(info_fake)

        info = {
            "context": str(data[i]['claim'])
        }
        results.append(info)

'''save results'''
with open('HOVER_supports_dev.json', 'w+') as f1:
    json.dump(results, f1)

with open('HOVER_supports_dev_fake.json', 'w+') as f2:
    json.dump(results_fake, f2)

print('Count：', count)
```
